# Data Cleaning in Python

```python
type(df["job_posted_date"][0])
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df["job_posted_month"] = df["job_posted_date"].dt.month # creating month number from the converted date time
```

```python
df.sort_values(by = "job_posted_date", ascending=False, inplace=True) #  sort job_posted_date deccending
```

### Delete un-use (salary_hour_avg) column

```python
df.drop(labels="salary_hour_avg", axis=1, inplace=True) 
```

### Delete all na or Not a number valeue from salary_year_avg column

```python
df.dropna(subset=['salary_year_avg'], inplace=True) 
```

### Fill not a number values with the median salary

```python
median_year_salary = df["salary_year_avg"].median()

df_filled = df;
df_filled["salary_year_avg"] = df_filled["salary_year_avg"].fillna(median_year_salary)
#df_filled.loc[:10, "salary_year_avg"]
```

### Remove Duplicate Values 

```python
df_unique = df_filled;
df_unique = df_unique.drop_duplicates()
df_unique

# to get the number of duplicated values Removed
print(len(df_filled))
print(len(df_unique))
print(len(df_filled) - len(df_unique))
```

# (2) Calculate Projected Salary Next Year

1 Senior roles assume 3%

2 Order roles assume 3%

```python
df_salary = df[pd.notna(df["salary_year_avg"])].copy()

def projected_salsry(row):
    if "Senior" in row["job_title_short"]:
        return 1.05 * row["salary_year_avg"]
    else:
       return 1.03 * row["salary_year_avg"] 
```

```python
df_salary["salary_year_inflated"] = df_salary.apply(projected_salsry, axis=1)
```

# (7) What are the most demanded skills for the top 3 most popular data roles ?

```python
df_US = df[df["job_country"] == "United States"]

#df_US["job_skills"] = df_US["job_skills"].apply(lambda x: x.strip("[]").replace("'", "").split(","))
df_US.loc[:, "job_skills"] = df_US["job_skills"].apply(lambda x: [] if pd.isna(x) else x.strip("[]").replace("'", "").split(", "))
df_skills = df_US.explode('job_skills')

#df_skills[["job_title_short","job_skills"]
df_skills_count = df_skills.groupby(["job_skills", "job_title_short"]).size()
df_skills_count = df_skills_count.reset_index(name = "skill_count")
df_skills_count.sort_values(by = "skill_count", ascending = False, inplace = True)
```

```python
job_titles = df_skills_count["job_title_short"].unique().tolist()
job_titles = sorted(job_titles[:3])
```

```python
df_job_title_count = df_US["job_title_short"].value_counts().reset_index(name="job_total")
df_skills_per = pd.merge(df_skills_count, df_job_title_count, how='left', on='job_title_short')
df_skills_per["skilll_percent"] = 100 *(df_skills_per["skill_count"] / df_skills_per["job_total"])
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

for i,job_title in enumerate(job_titles):
    df_plot = df_skills_per[df_skills_per["job_title_short"] == job_title].head(5)
    #df_plot.columns = ['job_skills', 'skilll_percent']
    # Plot the data
    sns.set_theme(style="ticks")
    sns.barplot(data=df_plot, x='skilll_percent', y='job_skills', ax=ax[i], palette='dark:b_r')
    ax[i].set_title(job_title)
    ax[i].set_ylabel("")
    ax[i].set_xlabel("")
    #ax[i].legend().set_visible(False)
    ax[i].set_xlim(0,78)
    ax[i].set_ylabel("")
    for n, v in enumerate(df_plot["skilll_percent"]):
        ax[i].text(v + 1, n, f'{v:.0f}%', va="center")
    ax[i].set_xticks([])    
fig.suptitle("Likelihood of Skills Requested in US Job Posting")    
fig.tight_layout(h_pad=0.5) # fix the overlap
plt.show()    
ax[i].legend_.remove() if ax[0].legend_ else None
ax[i].legend_.remove() if ax[1].legend_ else None
```

# (8) Trending Top Skills for Data Analyst in the US

```python
df_DA_US = df[(df["job_title"] == "Data Analyst") & (df["job_country"] == "United States")].copy()

df_DA_US["job_posted_month_no"] = df["job_posted_date"].dt.month

df_DA_US.loc[:, "job_skills"] = df_DA_US["job_skills"].apply(lambda x: [] if pd.isna(x) else x.strip("[]").replace("'", "").split(", "))
df_DA_US_EXPLODE = df_DA_US.explode('job_skills')
```

```python
df_DA_US_Pivot = df_DA_US_EXPLODE.pivot_table(
    index="job_posted_month_no",
    columns="job_skills",
    aggfunc="size",
    fill_value=0
)

df_DA_US_Pivot.loc["Total"] = df_DA_US_Pivot.sum()
df_DA_US_Pivot = df_DA_US_Pivot[df_DA_US_Pivot.loc["Total"].sort_values(ascending = False).index ]
df_DA_US_Pivot = df_DA_US_Pivot.drop("Total")

DA_Total = df_DA_US.groupby("job_posted_month_no").size()
df_DA_US_Percent = df_DA_US_Pivot.div(DA_Total/100, axis=0)

df_DA_US_Percent = df_DA_US_Percent.reset_index()
df_DA_US_Percent["job_posted_months"] = df_DA_US_Percent["job_posted_month_no"].apply(lambda x : pd.to_datetime(x,format="%m").strftime('%b'))

df_DA_US_Percent = df_DA_US_Percent.set_index('job_posted_months')
df_DA_US_Percent = df_DA_US_Percent.drop(columns="job_posted_month_no")
```

```python
df_plot = df_DA_US_Percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style="ticks")
sns.despine()

plt.xlabel('2023')
plt.ylabel('Likelihood in Job Posting')
plt.title('Trending Top Skills for Data Analyst in the US')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
plt.show()
```

# (11) What is the most Optimal skills to learn for Data Analysts ? 

```python
df_DA_US = df[(df["job_title_short"] == "Data Analyst") & (df["job_country"] == "United States")].copy()

df_DA_US = df_DA_US.dropna(subset="salary_year_avg")

```

```python
df_DA_US.loc[:, "job_skills"] = df_DA_US["job_skills"].apply(lambda x: [] if pd.isna(x) else x.strip("[]").replace("'", "").split(", "))
df_DA_US = df_DA_US.explode('job_skills')
#df_DA_US[["salary_year_avg", "job_skills"]].head(5)
```

```python
df_DA_skills = df_DA_US.groupby("job_skills")["salary_year_avg"].agg(["count","median"]).sort_values(by = "count", ascending = False)
df_DA_skills = df_DA_skills.rename(columns={"count":"skill_count","median":"median_salary"})
DA_job_Count = len(df_DA_US)
df_DA_skills["skill_percent"] = df_DA_skills["skill_count"] / DA_job_Count * 100

df_DA_skills = df_DA_skills.head(15)

df_DA_skills.plot(kind = "scatter", x = 'skill_count', y = 'median_salary')
plt.title("Most Optimal Skills for Data Analyst in US")
plt.xlabel("Percent of Data Analyst Job")
plt.ylabel("Median Yearly Salary (USD)")
fig.tight_layout()

for i, txt in enumerate(df_DA_skills.index):
    plt.text(df_DA_skills["skill_count"].iloc[i], df_DA_skills["median_salary"].iloc[i], txt)
    ax = plt.gca()
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f"${int(y/1000)}k"))
plt.show()
```
